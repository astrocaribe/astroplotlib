# making annotations on a FITS image

```python
# Import necessary modules
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pyfits as pf
```

```python
# Name output file
psname = 'python_plot.ps'

# Read input image and select subarray
image = pf.getdata('./data/u9w10107m_drz.fits', 1)
image = image[100:800, 100:800]

# Define coordinates of sources
xc = [112.40, 194.21, 172.80, 385.57, 152.96, 302.10, 317.19, 195.28, 389.43]
yc = [ 399.14, 443.14,  496.3, 407.87, 677.33, 670.35, 675.82, 473.09, 376.93]
```

## using plt.colorbar

```python
# Set up plot
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_xlabel('x pixels')
ax.set_ylabel('y pixels')
ax.set_xlim(0, 700)
ax.set_ylim(0, 700)

# Plot image and overplot sources
cmap = cm.get_cmap('jet')
img = ax.imshow(image, cmap=cmap, vmin=0, vmax=6, interpolation='bilinear')
ax.plot(xc, yc, marker='o', mec='white', mfc='none', ms=15, ls='none')

# Add a colorbar
ax_cb = fig.add_axes([0.2, 0.2, 0.3, 0.025])

cbar = plt.colorbar(mappable=img, cax=ax_cb, orientation='horizontal', ticks=[6, 4, 2, 0])
for i, j in enumerate(cbar.ax.spines): cbar.ax.spines[j].set_color('red')

cbar.ax.invert_xaxis()

# Save figure
fig.savefig(psname)
```
