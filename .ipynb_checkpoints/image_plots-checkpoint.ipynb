# displaying a FITS image

```python
# Invoke inline plotting
%pylab inline
```

```python
# Import neccesary modules
import pyfits as pf
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python
# Name output file
psname = 'image_plot_1.eps'

# Set up plot
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlabel('x pixels')
ax.set_ylabel('y pixels')

# Read in image and select subarray
image = pf.getdata('./data/ib6w71lxq_flt.fits', 4)
image = image[700:1400, 2000:2700]
tam = image.shape

# Generate plot
cmap = cm.get_cmap('gist_gray_r')
ax.imshow(image, cmap=cmap, vmin=0, vmax=100,
          interpolation='nearest')
ax.set_xlim(0, tam[1], plt.minorticks_on())
ax.set_ylim(0, tam[0], plt.minorticks_on())

# Display figure
display(fig)

# Save figure
fig.savefig(psname)
```

# display JPEG

```python
# Import necessary modules
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
# Name output file
psname = 'image_plot_2.eps'

# Import JPEG and flip top/bottom
image = Image.open('./data/soho_image.jpg')
image = np.array(image.transpose(1))

# Set up plot
fig = plt.figure(figsize=(6, 12))
ax_image = plt.axes([.1, .1, .8, .5], frameon=False)
ax_inset = plt.axes([.1, .57, .4, .2], frameon=False)
ax_image.set_axis_off()
ax_inset.set_axis_off()

# Display main image and inset location
ax_image.plot(850, 850, marker='s', mec='lime', mfc='none',ms=100, mew=1.2)
ax_image.imshow(image)

# Display inset
sub_image = image[700:1000, 700:1000, :]
ax_inset.plot(150, 150, marker='s', mec='lime', mfc='none', ms=172, mew=2)
ax_inset.imshow(sub_image)

# Display figure
display(fig)

# Save output file
fig.savefig(psname)
```

# making annotations on a FITS image

```python
# Import necessary modules
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pyfits as pf
```

```python
# Name output file
psname = 'image_plot_3.eps'

# Set up plot
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_xlabel('x pixels')
ax.set_ylabel('y pixels')
ax.set_xlim(0, 700)
ax.set_ylim(0, 700)

# Read input image and select subarray
image = pf.getdata('./data/u9w10107m_drz.fits', 1)
image = image[100:800, 100:800]

# Define coordinates of sources
xc = [112.40, 194.21, 172.80, 385.57, 152.96, 302.10, 317.19, 195.28, 389.43]
yc = [ 399.14, 443.14,  496.3, 407.87, 677.33, 670.35, 675.82, 473.09, 376.93]

# Plot image and overplot sources
cmap = cm.get_cmap('jet')
img = ax.imshow(image, aspect='auto', cmap=cmap, vmin=0, vmax=6, interpolation='bilinear')
ax.plot(xc, yc, marker='o', mec='white', mfc='none', ms=15, ls='none')

# Add a colorbar
ax_cb = fig.add_axes([0.2, 0.2, 0.3, 0.025])
# for i, j in enumerate(ax_cb.spines): ax_cb.spines[j].set_color('red')
    
cbar = plt.colorbar(img, cax=ax_cb, orientation='horizontal', ticks=[0, 2, 4, 6])
# cbar.ax.set_xlim(0, -200)
# cbar.ax.set_ylim(0, 200)
cbar.ax.invert_xaxis()

#cbar.ax.spines[:].set_color('white')
#ax_cb.set_axis_off()

cbar.solids.set_edgecolor('white')

#for i, j in enumerate(cbar.ax.spines): cbar.ax.spines[j].set_color('white')
#cbar.ax.draw_artist()
    

#for ax in enumerate(cbar.ax.spines):
#    #ax.tick_params(color='green', labelcolor='green')
#    for spine in ax.spines.values():
#        spine.set_edgecolor('green')    
    
# Display figure
display(fig)

# Save figure
fig.savefig(psname)
```

# colour tables

```python
# Import necessary modules
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pyfits as pf
import numpy as np
```

## without contour lines

```python
# Name output file
psname = 'image_plot_4-1.eps'

# Read FITS file
img = pf.getdata('./data/proplyd.fits', 0)

# Setup plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
ax1.set_aspect('equal')
ax2.set_aspect('equal')
ax1.set_xlabel(r'$\Delta \alpha$ ["]')
ax1.set_ylabel(r'$\Delta \delta$ ["]')
ax1.set_title('Disk 114-426')
ax2.set_title('345 GHz')
ax2.set_xlabel(r'$\Delta \alpha$ ["]')

# Plot left figure
cmap = cm.get_cmap('gist_heat_r')
ax1.imshow(img, cmap=cmap, vmin=0, vmax=6, extent=[4,-4,-4,4])

# Plot right figure
# Compute contours
contour_x = 4.0 - np.arange(90)* ((4.0 - (-4.0))/89.)
contour_y = -4.0 + np.arange(90)* ((4.0 - (-4.0))/89.)
ax2.set_xlim(4,-4)

# Set cmap colour scheme
cmap = cm.get_cmap('jet')

levels = [0.0, 0.6, 0.8, 0.9,1.0, 1.2, 2.0, 3.0, 5.0, 7.0]
cont = ax2.contourf(contour_x, contour_y, img, cmap=cmap, vmin=0.1, vmax=3, levels=levels)
ax2.ticklabel_format(axis='y', visible='False')

# Display figure
display(fig)

# Save figure
fig.savefig(psname)
```

## with contour lines

```python
# Name output file
psname = 'image_plot_4-2.eps'

# Read FITS file
img = pf.getdata('./data/proplyd.fits', 0)

# Setup plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
ax1.set_aspect('equal')
ax2.set_aspect('equal')
ax1.set_xlabel(r'$\Delta \alpha$ ["]')
ax1.set_ylabel(r'$\Delta \delta$ ["]')
ax1.set_title('Disk 114-426')
ax2.set_title('345 GHz')
ax2.set_xlabel(r'$\Delta \alpha$ ["]')

# Plot left figure
cmap = cm.get_cmap('gist_heat_r')
ax1.imshow(img, cmap=cmap, vmin=0, vmax=6, extent=[4,-4,-4,4])

# Plot right figure
# Compute contours
contour_x = 4.0 - np.arange(90)* ((4.0 - (-4.0))/89.)
contour_y = -4.0 + np.arange(90)* ((4.0 - (-4.0))/89.)
ax2.set_xlim(4,-4)

# Set cmap colour scheme
cmap = cm.get_cmap('jet')

levels = [0.0, 0.6, 0.8, 0.9,1.0, 1.2, 2.0, 3.0, 5.0, 7.0]
cont = ax2.contourf(contour_x, contour_y, img, cmap=cmap, vmin=0.1, vmax=3, levels=levels)

# Includes black contour lines
contour_lines = ax2.contour(contour_x, contour_y, img, colors=['black'], levels=levels)
plt.clabel(contour_lines, fontsize=6)

ax2.ticklabel_format(axis='y', visible='False')

# Display figure
display(fig)

# Save figure
fig.savefig(psname)
```
